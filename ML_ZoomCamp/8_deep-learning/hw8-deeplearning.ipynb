```python
import numpy as np
import tensorflow as tf

import warnings
warnings.filterwarnings('ignore')

SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)
```

```python
from tensorflow import keras
```

```python
# creating a sequential model
model = keras.Sequential()
```

```python
model.add(keras.layers.Conv2D(
    filters=32,
    kernel_size=(3,3),
    activation='relu',
    input_shape=(200,200,3)
))
```

```python
model.add(keras.layers.MaxPooling2D(
    pool_size=(2,2)
))
```

```python
model.add(keras.layers.Flatten())
```

```python
model.add(keras.layers.Dense(64, activation='relu'))
```

```python
model.add(keras.layers.Dense(1, activation='sigmoid'))
```

```python
optimizer = keras.optimizers.SGD(
    learning_rate=0.002,
    momentum=0.8
)
```

### Q1.

Since its binary classification problem, binary crossentropy works best

```python
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

### Q2.

```python
model.summary()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_gen = ImageDataGenerator(rescale=1./255)

train_ds = train_gen.flow_from_directory(
    '/kaggle/input/hair-types/data/train',
    target_size=(200,200),
    class_mode='binary',
    batch_size=20,
    shuffle=True
)
```

```python
test_gen = ImageDataGenerator(rescale=1./255)

test_ds = test_gen.flow_from_directory(
    '/kaggle/input/hair-types/data/test',
    target_size=(200,200),
    class_mode='binary',
    batch_size=20,
    shuffle=True
)
```

```python
history = model.fit(
    train_ds,
    epochs=10,
    validation_data=test_ds
)
```

```python
import matplotlib.pyplot as plt
```

### Q3.

```python
plt.plot(history.history['accuracy'], label='train_accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.legend()
plt.show()
```

```python
round(np.median(history.history['accuracy']),2)
```

### Q4.

```python
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend()
plt.show()
```

```python
round(np.std(history.history['loss']),3)
```
